        <h2><strong>مقدمه</strong></h2>
            در این پروژه، یک مدل یادگیری ماشین برای <strong>شناسایی تراکنش‌های جعلی</strong> توسعه داده می‌شود. سیستم‌های مالی امروزی نیاز به مدل‌هایی دارند که بتوانند تراکنش‌های مشکوک را در لحظه شناسایی کنند و از وقوع تقلب جلوگیری نمایند. هدف این پروژه، استفاده از الگوریتم‌های یادگیری ماشین برای ایجاد یک مدل قابل‌اتکا جهت شناسایی تراکنش‌های جعلی در بانکداری است.
        <h2><strong>توضیح دیتاست</strong></h2>
            مجموعه داده مورد استفاده شامل 20 ویژگی مختلف از تراکنش‌های بانکی است که شامل مشخصات حساب، نوع تراکنش، اطلاعات دستگاه، موقعیت جغرافیایی و سابقه تقلب می‌شود.
            <br>
            <h2><strong>وظیفه شما در این تمرین:</strong></h2>
            <li>✅ <strong>پیش‌پردازش داده‌ها:</strong> شامل تبدیل مقادیر دسته‌ای به عددی با استفاده از <code>LabelEncoder</code>، حذف مقادیر نامعتبر و نرمال‌سازی داده‌ها.</li>
            <li>✅ <strong>آموزش مدل:</strong> استفاده از مدل‌های یادگیری ماشین مانند <strong>GradientBoosting, LightGBM, AdaBoost، XGBoost</strong> برای تشخیص تقلب.</li>

            <li>✅ <strong>ذخیره و بارگذاری مدل:</strong> مدل آموزش‌دیده با فرمت <code>joblib</code> ذخیره و در برنامه Streamlit بارگذاری می‌شود.</li>

````python
# پر کردن با میانگین ستون
df['Daily_Transaction_Count'] = df['Daily_Transaction_Count'].fillna(
    df['Daily_Transaction_Count'].mean()
)

# پر کردن با میانه ستون 
df['Transaction_Amount'] = df['Transaction_Amount'].fillna(
    df['Transaction_Amount'].median()
)

# پر کردن با مُد 
df['Is_Weekend'] = df['Is_Weekend'].fillna(
    df['Is_Weekend'].mode()[0]
)

df['Account_Balance'] = df['Account_Balance'].fillna(
    df['Account_Balance'].mode()[0]
)

df['Avg_Transaction_Amount_7d'] = df['Avg_Transaction_Amount_7d'].fillna(
    df['Avg_Transaction_Amount_7d'].mode()[0]
)

df['Failed_Transaction_Count_7d'] = df['Failed_Transaction_Count_7d'].fillna(
    df['Failed_Transaction_Count_7d'].mode()[0]
)

df['Transaction_Distance'] = df['Transaction_Distance'].fillna(
    df['Transaction_Distance'].mode()[0]
)

df['Risk_Score'] = df['Risk_Score'].fillna(
    df['Risk_Score'].mode()[0]
)


df['Fraud_Label'] = df['Fraud_Label'].fillna(
    df['Fraud_Label'].mode()[0]
)


df['Previous_Fraudulent_Activity'] = df['Previous_Fraudulent_Activity'].fillna(
    df['Previous_Fraudulent_Activity'].mode()[0]
)

df['IP_Address_Flag'] = df['IP_Address_Flag'].fillna(
    df['IP_Address_Flag'].mode()[0]
)

df['Transaction_Amount'] = df['Transaction_Amount'].fillna(
    df['Transaction_Amount'].mode()[0]
)

df['Card_Age'] = df['Card_Age'].fillna(
    df['Card_Age'].mode()[0]
)

# df['Transaction_Type'] = df['Transaction_Type'].fillna(
#     df['Transaction_Type'].mode()[0]
# )

# df['Transaction_Amount_Quantized'] = df['Transaction_Amount_Quantized'].fillna(
#     df['Transaction_Amount_Quantized'].mode()[0]
# )


# df['Account_Balance_Quantized'] = df['Account_Balance_Quantized'].fillna(
#     df['Account_Balance_Quantized'].mode()[0]
# )


 
````

````python
def hist_plot(df, numeric_columns, bins=12, rot=None, figsize=(14, 8)):
    sns.set_style("darkgrid")
    num_vars = len(numeric_columns)
    rows = (num_vars // 3) + (num_vars % 3 > 0)  
    cols = min(3, num_vars)

    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = axes.flatten()  
    for i, col in enumerate(numeric_columns):
        ax = axes[i]  
        sns.histplot(data=df, x=col, bins=bins, ax=ax, kde=True)  # رسم هستوگرام
        ax.set_title(f'Distribution {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')
        if rot:
            ax.set_xticklabels(ax.get_xticklabels(), rotation=rot)

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    plt.tight_layout()
    plt.show()
````

````python
# TODO: شناسایی داده های پرت به کمک IQR
import pandas as pd
import numpy as np

Q1 = df['Transaction_Amount'].quantile(0.25)
Q3 = df['Transaction_Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Transaction_Amount'] < lower_bound) | (df['Transaction_Amount'] > upper_bound)]

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"تعداد داده‌های پرت: {len(outliers)}")
print("نمونه داده‌های پرت:")
print(outliers[['IP_Address_Flag']].head())
````

````python
# TODO: شناسایی داده های پرت به کمک Z-Score
import pandas as pd
import numpy as np
from scipy import stats

mean_amount = df['Transaction_Amount'].mean()
std_amount = df['Transaction_Amount'].std()

df['Z_Score'] = np.abs(stats.zscore(df['Transaction_Amount']))

outliers = df[df['Z_Score'] > 3]

print(f"میانگین مبلغ تراکنش: {mean_amount:.2f}")
print(f"انحراف معیار مبلغ تراکنش: {std_amount:.2f}")
print(f"تعداد داده‌های پرت: {len(outliers)}")
print("نمونه داده‌های پرت:")
print(outliers[['Transaction_Amount', 'Z_Score']].head())
````

        <br><br>
        1. <strong>ورودی‌ها:</strong>
            <li><strong>df:</strong> دیتافریم ورودی که شامل داده‌های اصلی شما است و ممکن است برخی مقادیر گمشده داشته باشد.</li>

            <li><strong>auto_impute:</strong> یک پارامتر بولی که اگر مقدار آن <code>True</code> باشد، داده‌ها به طور خودکار پر می‌شوند. برای ستون‌های عددی با میانگین و برای ستون‌های رشته‌ای با مد پر می‌شود. در غیر این صورت، از <code>strategy_dict</code> استفاده می‌شود.</li>
        <br>
        2. <strong>پر کردن خودکار (اگر <code>auto_impute=True</code>):</strong>
            <li>در این بخش، اگر شما <code>auto_impute</code> را <code>True</code> قرار دهید، تمامی ستون‌های عددی به صورت خودکار با میانگین پر می‌شوند و ستون‌های رشته‌ای با مد پر خواهند شد. این روش سریع و راحت است اما ممکن است برای همه داده‌ها مناسب نباشد.</li>
        <br>
        3. <strong>پر کردن با استراتژی مشخص (اگر <code>strategy_dict</code> داده شده):</strong>

````python
# TODO :  بررسی  عملکرد مدل روی داده های آموزش و تستس

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, 
                            recall_score, f1_score, roc_auc_score,
                            confusion_matrix, classification_report)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

y_train_proba = model.predict_proba(X_train)[:, 1]  
y_test_proba = model.predict_proba(X_test)[:, 1]

metrics = {
    'Accuracy': [accuracy_score(y_train, y_train_pred), accuracy_score(y_test, y_test_pred)],
    'Precision': [precision_score(y_train, y_train_pred), precision_score(y_test, y_test_pred)],
    'Recall': [recall_score(y_train, y_train_pred), recall_score(y_test, y_test_pred)],
    'F1-Score': [f1_score(y_train, y_train_pred), f1_score(y_test, y_test_pred)],
    'ROC-AUC': [roc_auc_score(y_train, y_train_proba), roc_auc_score(y_test, y_test_proba)]
}

results_df = pd.DataFrame(metrics, index=['Train', 'Test'])
print("نتایج ارزیابی:")
print(results_df)

print("\nگزارش طبقه‌بندی برای داده تست:")
print(classification_report(y_test, y_test_pred))

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_train, y_train_pred), annot=True, fmt='d', cmap='Blues')
plt.title('ماتریس درهم‌ریختگی (آموزش)')

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test, y_test_pred), annot=True, fmt='d', cmap='Blues')
plt.title('ماتریس درهم‌ریختگی (تست)')
plt.tight_layout()
plt.show()

train_test_diff = (results_df.loc['Train'] - results_df.loc['Test']).abs()
print("\nتفاوت بین عملکرد آموزش و تست:")
print(train_test_diff)

if train_test_diff.mean() > 0.15:  # اگر میانگین تفاوت بیشتر از 15% باشد
    print("\n⚠️ هشدار: ممکن است مدل دچار بیش‌برازش شده باشد")
else:
    print("\n✅ مدل عملکرد متعادلی روی داده‌های آموزش و تست دارد")
````
